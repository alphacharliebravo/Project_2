```python
# importing libraries, dependencies
import pandas as pd 
import datetime as DT
import numpy as np
import os 
pd.options.mode.chained_assignment = None
```

```python
# reading csv files into dataframes
df1 = pd.read_csv("./Resources/fake_data_address.csv", dtype={'first_name': str, 'middle_name': str, 'last_name': str, 'zip':str})
df2 = pd.read_csv("./Resources/fake_data_dob.csv", dtype={'last_name, first_name middle_name': str, 'dob': str, 'email': str, 'guid':str})
```

```python
# display dataframe 1
df1.head()
```

```python
# display dataframe 2
df2.head()
```

```python
# drop empty rows
df1 = df1.dropna()
df2 = df2.dropna()
```

```python
# correct zip code errors
df1['zip'] = df1['zip'].astype(int).astype(str).str.zfill(5)
df1.head()
```

```python
# splitting name column into last name column and first name + middle name column
df2[['last', 'f_m']] = df2['last_name, first_name middle_name'].apply(lambda x: pd.Series(str(x).split(", ")))
df2.head()
```

```python
# drop original name column
df2 = df2.drop('last_name, first_name middle_name', axis=1)
```

```python
# combine first name and middle name column
df1['f_m'] = df1.agg('{0[first_name]} {0[middle_name]}'.format, axis=1)
df1
```

```python
# merge two data frame by first name + middle name column
df = df1.merge(df2, on='f_m', how='inner')
df.head()
```

```python
# drop columns with less important data
df = df.drop('guid', axis=1)
df = df.drop('f_m', axis=1)
df = df.drop('last', axis=1)
df = df.drop('id', axis=1)
df.head(5)
```

```python
# convert date of birth column from string to datetime
df['dob'] = pd.to_datetime(df['dob'])
df
```

```python
# calculate age
today = pd.Timestamp('2023-03-20 00:00:00')
df['dob'] = pd.to_datetime(df['dob'], format='%Y-%m-%d')  
df['dob'] = df['dob'].where(df['dob'] < today, df['dob'] -  np.timedelta64(100, 'Y')) 
df['age'] = (today - df['dob']).astype('<m8[Y]')  
df.head()
```

```python
# filter people with 40=<age=<60
df = df.loc[(df['age'] >= 40) & (df['age'] <= 60)]
```

```python
# separate address column into Line1 column and Line2 column
df[['Line1', 'Line2']] = df['Line 1: address Line 2: address2'].apply(lambda x: pd.Series(str(x).split(" Line 2:")))
df[['delete', 'Line1']] = df['Line1'].apply(lambda x: pd.Series(str(x).split("Line 1:")))
df = df.drop('Line 1: address Line 2: address2', axis=1)
df = df.drop('delete', axis=1)
df.head()
```

```python
# separate state-city column into state column and city column, drop original column(state-city)
df[['State', 'City']] = df['state-city'].apply(lambda x: pd.Series(str(x).split("-")))
df = df.drop('state-city', axis=1)
df.head()
```

```python
# create address colum
for address in df['Line2']:
    if  (address== ''):
        df['address'] = df.agg('{0[Line1]}, {0[City]}, {0[State]}, {0[zip]}'.format, axis=1)
    else:
        df['address'] = df.agg('{0[Line1]}, {0[Line2]},{0[City]}, {0[State]}, {0[zip]}'.format, axis=1)
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df['name']= df.apply(lambda row: ' '.join([str(row['first_name']), str(row['middle_name']), str(row['last_name'])]), axis=1)

# Loop through each row in the dataframe
for index, row in df.iterrows():
    mailing_name=row['name']
    address=row['address']
#format address
    line1, line2, city_state_zip = address.split(',', 2)
    city, state_zip = city_state_zip.split(',', 1)
    state, zip_code = state_zip.split(',', 1)
    formatted_address = f"{line1}\n{line2}\n{city}, {state} {zip_code}"
# Combine the values from col1 and col2, separated by a comma
    mailing_value = f"{ mailing_name}\n{formatted_address}\n"
# Print each combined value on a new line
    print(mailing_value)
```
